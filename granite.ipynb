```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-3.3-8b-instruct")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-3.3-8b-instruct")
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_id = "ibm-granite/granite-3.3-8b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",  # Automatically uses GPU/CPU
    torch_dtype=torch.float16  # For performance
)

granite_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

```

```python
def ask_granite(prompt):
    result = granite_pipeline(
        prompt,
        max_length=300,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.95
    )
    return result[0]['generated_text']

# Example
response = ask_granite("What are 3 simple ways to manage daily expenses?")
print(response)
```

```python
print(ask_granite("How to start investing with a small salary?"))
print(ask_granite("Give me a daily savings plan for students."))
print(ask_granite("Explain the 50/30/20 budget rule."))
```

```python
# Install
!pip install transformers accelerate sentencepiece

# Import
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model and tokenizer
model_name = "ibm-granite/granite-3.3-8b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", trust_remote_code=True)

# Define function
def ask_granite(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_length=300)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Ask multiple questions
print(ask_granite("How to save money on monthly expenses?"))
print(ask_granite("Explain different types of investment options."))
print(ask_granite("How to build a credit score?"))
print(ask_granite("Explain tax-saving strategies for students."))
print(ask_granite("How to plan a retirement fund starting at age 25?"))
```
